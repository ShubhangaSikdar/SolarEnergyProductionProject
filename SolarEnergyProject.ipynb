```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the csv file into a DataFrame
data = pd.read_csv('hourly-weather-dataset_random-order.csv')
data.head() # Displaying 5 rows of the data.
```

```python
# Drop duplicate rows from data
data.drop_duplicates(inplace=True)
data
```

```python
# Display the number of missing values in each column
data.isnull().sum()
```

```python
df = data.dropna()
df
```

```python
for col in data.columns:
  if col[:1] == '01':
    data.rename(columns={col : 'id' + col[4:]}, inplace = True)
data.head()
```

```python
# Check if the number of unique values in the 'Random' column equals the total number of rows
if data['Random'].nunique() == len(data):
    print("All values in the column are unique.")
else:
    print("There are duplicate values in the column.")
```

```python
# Reset the index of the DataFrame
df = data.reset_index()
```

```python
# Set index to 'Random'
df = data.set_index('Random')
df
```

```python
# Sort the DataFrame by its index
df = data.sort_index()
df
```

```python
# Fill the missing values in the DataFrame using forward fill method
df = data.fillna(method = 'ffill')
df.head()
```

```python
# Convert all columns in the DataFrame to numeric types, coercing errors to NaN.
data = data.apply(pd.to_numeric, errors='coerce')
print(data)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()  # Create an Instance of MinMaxScaler
normalized_data = scaler.fit_transform(data)  # Fit the scaler to the data and transform it to a range between 0 and 1
normalized_data = pd.DataFrame(normalized_data, columns=data.columns)  # Convert the normalized data back to DataFrame
print(normalized_data.head())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()   # Create an Instance of StandardScaler
standardized_data = scaler.fit_transform(data)  # Fit the scaler to the data and transform it to have a mean of 0 and a standard deviation of 1
standardized_data = pd.DataFrame(standardized_data, columns=data.columns)   # Convert the standardized data back to dataFrame
print(standardized_data.head())
```
